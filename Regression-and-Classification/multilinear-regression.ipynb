```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('winequality-red.csv')
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
X = df.loc[:, df.columns != 'quality'] # a pandas dataframe
X_vals = X.values # a 2d numpy array (aka a matrix)
y = df['quality'] # a pandas series
y_vals = y.values # a 2d matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_vals, y_vals, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```python
coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient'])  
coeff_df
```

```python
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

df.head(25)
```

```python
regressor.score(X_vals, y_vals)
```

```python
regressor.score(X_train, y_train)
```
