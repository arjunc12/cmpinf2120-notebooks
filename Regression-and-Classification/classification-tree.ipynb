```python
from sklearn.tree import DecisionTreeClassifier 
```

```python
import pandas as pd

pima = pd.read_csv("diabetes.csv")
X = pima.loc[:, pima.columns != 'Outcome']
y = pima['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_Hat = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(Y_test, Y_Hat)
```

```python
from sklearn import tree
import pylab

fig = pylab.figure(figsize=(25,20))
_ = tree.plot_tree(model,
                   filled=True)
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_leaf_nodes = 10)
dt = model.fit(X_train, Y_train)


Y_Hat = model.predict(X_test)
accuracy_score(Y_test, Y_Hat)
```

```python
from sklearn import tree
import pylab

fig = pylab.figure(figsize=(50,40))
_ = tree.plot_tree(model,
                   filled=True,
                  feature_names = X.columns,
                  class_names = pima['Outcome'].unique().astype(str))
```

```python
import graphviz
```

```python
dot_data = tree.export_graphviz(model, out_file=None, 
                                feature_names=X.columns,  
                                class_names=pima['Outcome'].unique().astype(str),
                                filled=True)
```

```python
graph = graphviz.Source(dot_data, format="png") 
graph
```

```python
viz = dtreeviz(model, X, y,
                target_name="target",
                feature_names=X.columns,
                class_names=list(pima['Outcome'].unique()))

viz
```
