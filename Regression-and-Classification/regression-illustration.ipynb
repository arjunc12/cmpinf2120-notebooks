```python
import numpy as np
import seaborn as sns
sns.set()
import pylab
%matplotlib inline
```

```python
x = np.arange(0, 12, 2)
y = x * 2 + 5  + np.random.normal(0, 2, len(x))
```

```python
pylab.scatter(x, y)
pylab.savefig('toy-data.png', format='png')
```

```python
yhat1 = 0.75 * x + 15
pylab.scatter(x, y)
pylab.plot(x, yhat1)
for p1, p2, p3 in zip(x, y, yhat1):
    pylab.plot([p1, p1], [p2, p3], color='r', linewidth=4)
pylab.savefig('toy-data-bad-line.png', format='png')
```

```python
yhat2 = 2 * x + 5
pylab.scatter(x, y)
pylab.plot(x, yhat2)
for p1, p2, p3 in zip(x, y, yhat2):
    pylab.plot([p1, p1], [p2, p3], color='r', linewidth=4)
pylab.savefig('toy-data-correct-line.png', format='png')
```

```python
xnew = np.arange(1, 12, 2)
y_pred = 2 * xnew + 5
pylab.scatter(x, y)
pylab.plot(x, yhat2)
pylab.scatter(xnew, y_pred, color='g')
pylab.savefig('toy-data-predictions.png', format='png')
```

```python
pylab.scatter(x, y)
ybar = y.mean()
pylab.hlines(ybar, min(x), max(x), color='k')
pylab.plot(x, yhat2)
for i, (p1, p2, p3) in enumerate(zip(x, y, yhat2)):
    label1 = None
    label2 = None
    if i == 0:
        label1 = 'SSR'
        label2 = 'SST'
    else:
        label1 = '_nolegend_'
        label2 = '_nolegend_'    
    pylab.plot([p1, p1], [ybar, p3], color='r', linewidth=4, label=label1)
    pylab.plot([p1, p1], [ybar, p2], color='g', linewidth=6, alpha=0.5, label=label2)
    
pylab.legend()
pylab.savefig('toy-data-rsquared.png', format='png')
```
