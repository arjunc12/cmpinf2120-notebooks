```python
import numpy as np
import pylab
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x = np.tile(np.arange(-20, 20), 2)
x = x + np.random.normal(0, 0.2, size=len(x))
y = 2 + 2 * x - 0.5 * (x ** 2) + np.random.normal(loc=0, scale=50, size=len(x))
pylab.scatter(x, y)
```

```python
X = np.array([x]).T
Y = np.array([y]).T
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
pylab.subplot(2, 1, 1)
Y_hat_train = model.predict(X_train)
print(model.score(X_train, Y_train))
pylab.scatter(X_train[:, 0], Y_train)
pylab.scatter(X_train[:, 0], Y_hat_train)

pylab.subplot(2, 1, 2)
Y_hat_test = model.predict(X_test)
print(model.score(X_test, Y_test))
pylab.scatter(X_test[:, 0], Y_test)
pylab.scatter(X_test[:, 0], Y_hat_test)
```
