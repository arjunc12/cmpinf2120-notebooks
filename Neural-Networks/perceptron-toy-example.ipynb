```python
from sklearn.datasets import make_blobs
```

```python
blobs = make_blobs(centers=2)
features, classes = blobs
```

```python
import seaborn as sns
sns.scatterplot(x=features[:, 0], y=features[:, 1], c=classes)
```

```python
from sklearn.linear_model import Perceptron
perceptron = Perceptron()
perceptron.fit(features, classes)
perceptron.intercept_
```

```python
perceptron.coef_
```

```python
import pylab
x = pylab.linspace(features[:, 0].min(), features[:, 0].max())
```

```python
y = (-perceptron.intercept_[0] - (perceptron.coef_[0][0] * x)) / perceptron.coef_[0][1]
```

```python
sns.scatterplot(x=features[:, 0], y=features[:, 1], c=classes)
sns.lineplot(x=x, y=y)
```
